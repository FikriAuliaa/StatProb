```python
"""
FIKRI AULIA AS SA'ADI
5027231026
STATISTIKA DAN PROBABILITAS
"""

import statistics
import matplotlib.pyplot as plt

# DATA TINGGI BADAN SISWA KELAS A
data_tinggi_badan = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# FUNGSI UNTUK MENGHITUNG MEAN
def hitung_mean(data):
    return sum(data) / len(data)

# FUNGSI UNTUK MENGHITUNG MEDIAN
def hitung_median(data):
    data.sort()
    n = len(data)
    if n % 2 == 0:
        median = (data[n//2 - 1] + data[n//2]) / 2
    else:
        median = data[n//2]
    return median

# FUNGSI UNTUK MENGHITUNG MODUS
def hitung_modus(data):
    modus = max(set(data), key = data.count)
    return modus

# MENGHITUNG MEAN SECARA MANUAL
mean_manual = hitung_mean(data_tinggi_badan)

# MENGHITUNG MEDIAN SECARA MANUAL
median_manual = hitung_median(data_tinggi_badan)

# MENGHITUNG MODUS SECARA MANUAL
modus_manual = hitung_modus(data_tinggi_badan)

# MENGHITUNG MEAN, MEDIAN, DAN MODUS DENGAN LIBRARY STATISTICS
mean_stats = statistics.mean(data_tinggi_badan)
median_stats = statistics.median(data_tinggi_badan)
modus_stats = statistics.mode(data_tinggi_badan)

# MENAMPILKAN HASIL
print(f"Mean (Manual): {mean_manual}")
print(f"Mean (statistics): {mean_stats}")
print(f"Median (Manual): {median_manual}")
print(f"Median (statistics): {median_stats}")
print(f"Modus (Manual): {modus_manual}")
print(f"Modus (statistics): {modus_stats}")

# VISUALISASI DISTRIBUSI DATA
plt.hist(data_tinggi_badan, bins=5, edgecolor='black')
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Distribusi Tinggi Badan Siswa Kelas A')
plt.show()
```
